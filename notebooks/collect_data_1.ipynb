# I. Collecting dataset

### Import cÃ¡c thÆ° viá»‡n cáº§n thiáº¿t

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Má»Ÿ file `company_name.txt` Ä‘á»ƒ láº¥y danh sÃ¡ch tÃªn cÃ¡c cÃ´ng ty

```python
with open('../Data/company_name.txt', 'r') as file:
    companies = file.read().splitlines()
```

```python
companies[:5]
```

Thá»­ request cho cÃ´ng ti Ä‘áº§u tiÃªn

```python
URL = 'https://companiesmarketcap.com/apple/marketcap/'
response = requests.get(URL)

if response.ok:
    message = 'The request was successfully received.'
else:
    message = 'The request was not successful.'
message
```

Giá» ta sáº½ láº¥y thÃ´ng tin vá» `Country` vÃ  `Categories` cá»§a cÃ¡c cÃ´ng ty

```python
soup = BeautifulSoup(response.content, 'lxml')
soup.prettify
```

XÃ¡c Ä‘á»‹nh Ä‘Æ°á»£c cÃ¡c thÃ´ng tin cáº§n tÃ¬m trong ná»™i dung `response`

```python
table_container = soup.find('div', class_='table-container')
name = table_container.find('div', class_='company-name').text.strip()
market_cap = table_container.find(text='Marketcap').find_previous('div', class_='line1').text.strip()
country = table_container.find(text='Country').find_previous('div', class_='line1').text.strip()
share_price = table_container.find(text='Share price').find_previous('div', class_='line1').text.strip()
change_1_day = table_container.find(text='Change (1 day)').find_previous('div', class_='line1').text.strip()
change_1_year = table_container.find(text='Change (1 year)').find_previous('div', class_='line1').text.strip()


categories_div = table_container.find('div', class_='categories-box')
categories = [tag.text.strip() for tag in categories_div.find('div', class_='line1').find_all('a')]
categories_str = ', '.join(categories)

print(f"Name: {name}")
print(f"Marketcap: {market_cap}")
print(f"Country: {country}")
print(f"Share price: {share_price}")
print(f"Change (1 day): {change_1_day}")
print(f"Change (1 year): {change_1_year}")
print(f"Categories: {categories_str}")
```

BÃ¢y giá» tiáº¿n hÃ nh lÃ m tÆ°Æ¡ng tá»± vá»›i táº¥t cáº£ cÃ´ng ti

```python
all_data = {
    'Name': [],
    'Marketcap': [],
    'Country': [],
    'Share price': [],
    'Change (1 day)': [],
    'Change (1 year)': [],
    'Categories': []
}

for company in companies:
    URL = f'https://companiesmarketcap.com/{company}/marketcap/'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:92.0) Gecko/20100101 Firefox/92.0'}
    response = requests.get(URL, headers=headers)

    if response.ok:
        soup = BeautifulSoup(response.content, 'html.parser')
        print(company)
        table_container = soup.find('div', class_='table-container')
        name = table_container.find('div', class_='company-name').text.strip()
        market_cap = table_container.find(text='Marketcap').find_previous('div', class_='line1').text.strip()
        country = table_container.find(text='Country').find_previous('div', class_='line1').text.strip()
        share_price = table_container.find(text='Share price').find_previous('div', class_='line1').text.strip()
        change_1_day = table_container.find(text='Change (1 day)').find_previous('div', class_='line1').text.strip()
        change_1_year = table_container.find(text='Change (1 year)').find_previous('div', class_='line1').text.strip()

        categories_div = table_container.find('div', class_='categories-box')
        if categories_div:  
            categories = [tag.text.strip() for tag in categories_div.find('div', class_='line1').find_all('a')]
            categories_str = ', '.join(categories)
        else:
            categories_str = 'N/A' 
        # Append data to respective lists
        all_data['Name'].append(name)
        all_data['Marketcap'].append(market_cap)
        all_data['Country'].append(country)
        all_data['Share price'].append(share_price)
        all_data['Change (1 day)'].append(change_1_day)
        all_data['Change (1 year)'].append(change_1_year)
        all_data['Categories'].append(categories_str)

```

Táº¡o DataFrame `companies_df`

```python
# LÆ°u vÃ  dataframe
companies_df = pd.DataFrame(all_data)
```

```python
# LÆ°u vÃ o file csv
companies_df.to_csv('../Data/companies.csv', index=False)
```

ThÃ´ng qua kiá»ƒm tra cÃ¡c báº£ng khÃ¡c, ta nháº­n tháº¥y sá»‘ nÄƒm luÃ´n liÃªn tiáº¿p nhau.  
LÃ m tiáº¿p cho táº¥t cáº£ cÃ´ng ty

```python
all_company_data = []

for company in companies:
    URL = f'https://companiesmarketcap.com/{company}/marketcap/'
    response = requests.get(URL)
    
    if response.ok:
        soup = BeautifulSoup(response.content, 'html.parser')
        table = soup.find('tbody')
        if table:
            trs = table.find_all('tr')
            years = []
            data = []

            for tr in trs:
                td_elements = tr.find_all('td')
                if len(td_elements) == 3:
                    year = td_elements[0].text.strip()
                    years.append(year)
                    data.append(td_elements[1].text.strip())

            company_data = [company] + data
            all_company_data.append(company_data)


columns = ['Company'] + years
marketcap_df = pd.DataFrame(all_company_data, columns=columns)
```

```python
# LÆ°u vÃ o file csv
marketcap_df.to_csv('../Data/marketcap.csv', index=False)
```
