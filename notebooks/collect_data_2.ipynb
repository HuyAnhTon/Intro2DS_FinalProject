Ta th·∫•y r·∫±ng:

- C√°c t√™n c√¥ng ti trong c·ªôt `Company` kh√¥ng tr√πng v·ªõi t√™n c√¥ng ty trong c·ªôt `Name` trong `companies_df`.  

- Gi√° tr·ªã `2023 (TTM)` ƒë∆∞·ª£c ghi thay v√¨  `2023` trong `eps_df`, `earings_df`, `revenue_df`

```python
# Thay l·∫°i t√™n c√°c c√¥ng ty
# ƒê·∫ßu ti√™n ta t·∫°o m·ªôt dictionary v·ªõi key l√† gi√° tr·ªã t√™n c·∫ßn thay v√† value l√† t√™n s·∫Ω thay
companies_dict = dict(zip(comapanies_names[0].tolist(), companies_df['Name'].tolist()))
companies_dict
```

```python
# Thay l·∫°i t√™n c√°c c√¥ng ty
data_frames = [marketcap_df, revenue_df, earnings_df, eps_df, debts_df, net_assets_df, total_assets_df, cash_on_hand_df]
for df in data_frames:
    df['Company'].replace(companies_dict, inplace=True)

```

```python
# ƒê·ªïi l·∫°i t√™n c·ªôt
companies_df = companies_df.rename(columns={'Name': 'Company',})
eps_df = eps_df.rename(columns={'2023 (TTM)': '2023',})
earnings_df = earnings_df.rename(columns={'2023 (TTM)': '2023',})
revenue_df = revenue_df.rename(columns={'2023 (TTM)': '2023',})
```

Ta s·∫Ω merge c√°c DataFrame nh·ªè th√†nh m·ªôt DataFrame l·ªõn, v·ªõi c√°c c·ªôt l√†:

- `Company`

- `Year`

- `EPS`

- `Earnings`

- `Revenue`

- `Marketcap`

- `Total Debts`

- `Net Assets`

- `Total Assets`

- `Cash on hand`

- `Country`

- `Categories`

ƒê·ªÉ l√†m ƒë∆∞·ª£c ƒëi·ªÅu ƒë√≥, tr∆∞·ªõc ti√™n ta ph·∫£i `melt` c√°c `dataframe` hi·ªán c√≥ 

```python
eps_melted_df = eps_df.melt(id_vars='Company', var_name='Year', value_name='EPS')
earnings_melted_df = earnings_df.melt(id_vars='Company', var_name='Year', value_name='Earnings')
revenue_melted_df = revenue_df.melt(id_vars='Company', var_name='Year', value_name='Revenue')
marketcap_melted_df = marketcap_df.melt(id_vars='Company', var_name='Year', value_name='Marketcap')

debts_melted_df = debts_df.melt(id_vars='Company', var_name='Year', value_name='Total Debts')
net_assets_melted_df = net_assets_df.melt(id_vars='Company', var_name='Year', value_name='Net Assets')
total_assets_melted_df = total_assets_df.melt(id_vars='Company', var_name='Year', value_name='Total Assets')
cash_on_hand_melted_df = cash_on_hand_df.melt(id_vars='Company', var_name='Year', value_name='Cash on hand')
```

```python
cash_on_hand_melted_df.head(10)
```

```python
# Merge c√°c dataframe l·∫°i
merged_df = pd.merge(eps_melted_df, earnings_melted_df, on=['Company', 'Year'], how='outer')
merged_df = pd.merge(merged_df, revenue_melted_df, on=['Company', 'Year'], how='outer')
merged_df = pd.merge(merged_df, marketcap_melted_df, on=['Company', 'Year'], how='outer')

merged_df = pd.merge(merged_df, debts_melted_df, on=['Company', 'Year'], how='outer')
merged_df = pd.merge(merged_df, net_assets_melted_df, on=['Company', 'Year'], how='outer')
merged_df = pd.merge(merged_df, total_assets_melted_df, on=['Company', 'Year'], how='outer')
merged_df = pd.merge(merged_df, cash_on_hand_melted_df, on=['Company', 'Year'], how='outer')

merged_df = pd.merge(merged_df, companies_df,on=['Company'], how='outer' )
merged_df.sample(10)
```
